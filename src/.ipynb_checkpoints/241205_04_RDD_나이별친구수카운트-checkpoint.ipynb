### fakefriends.csv

- user_id : 사용자 ID
- name : 사용자 이름
- age : 사용자 나이
- num_friends : 친구 수


1. 평균 친구수 구하기
2. 가장 친구가 많은 사용자 찾기
3. 친구 수가 100 이상인 사용자 필터링
4. 특정 나이에 속한 사람들의 친구 수 합계
5. 분석 추가 주제

```python
from pyspark import SparkConf, SparkContext
import os

conf = SparkConf().setMaster('local').setAppName('friends')
sc = SparkContext(conf = conf).getOrCreate()
```

```python
sc
```

```python
direc = os.path.join(os.getcwd(), 'data')
file_name = 'fakefriends.csv'
file_path = os.path.join(direc, file_name)
file_path
```

```python
datas = sc.textFile('file:///' + file_path.replace('\\','/'))
datas.take(10)
```

```python
def parse(row):
    fields = row.split(',')
    user_id = fields[0]
    user_name = fields[1]
    user_age = fields[2]
    user_friends = fields[3]

    return (user_id, user_name, user_age, user_friends)

rows = datas.map(parse) #map은 transformation이기 때문에, 함수만 전달하면 됨
rows.take(10)
```

### 평균 친구수 구하기

```python
avg_friends = rows.map(lambda x : int(x[3])).sum() / \
rows.map(lambda x : x[3]).count()

print(f'평균 친구 수 입니다. {avg_friends:.1f}명')
```

### 가장 친구가 많은 사용자 찾기

```python
sort_friends_count = rows.map(lambda x : (x[1], int(x[3]))) \
                    . sortBy(lambda x : x[1], ascending = False).collect()

max_friends = sort_friends_count[0][1] #1등 저장해놓기

for num, (i, j) in enumerate(sort_friends_count):
    if max_friends == j:
        print(f'No.{num} 1등 명단 : {i}은 {j}명의 친구가 있습니다.')
    else:
        pass

    
```

### 친구 수가 100 이상인 사용자 필터링

```python
rows.filter(lambda x : int(x[3])>=100).sortBy(lambda x : int(x[3]), ascending = False).collect()
```

### 특정 나이에 속한 사람들의 친구 수 합계

```python
# 그냥 나이 전체 돌기
for i in rows.map(lambda x :int(x[2])).distinct().sortBy(lambda x : x).collect():
    try : 
        print(f'''
        {'*'*100}
        {str(i)}살의 친구 합계 : {rows.filter(lambda x : (int(x[2])) == i).map(lambda x : int(x[3])).sum()}명
        {str(i)}살의 친구 평균 : {rows.filter(lambda x : (int(x[2])) == i).map(lambda x : int(x[3])).sum() /
        rows.filter(lambda x : (int(x[2])) == i).map(lambda x : int(x[3])).count():.1f}명
        {'*'*100}''')
    except ZeroDivisionError as e :
        print(f'이 나이가 없나봐요 {e}')
    
    
```

```python
#나이대 추출(10대, 20대, ~~)
rows.map(lambda x :int(x[2]) // 10).distinct().sortBy(lambda x : x).collect()
```

```python
#10대부터 60대까지 전체 돌기
for i in rows.map(lambda x :int(x[2]) // 10).distinct().sortBy(lambda x : x).collect():
    print(f'''
    {'*'*100}
    {str(i)+'0'}대의 친구 합계 : {rows.filter(lambda x : (int(x[2])//10) == i).map(lambda x : int(x[3])).sum()}명
    {str(i)+'0'}대의 친구 평균 : {rows.filter(lambda x : (int(x[2])//10) == i).map(lambda x : int(x[3])).sum() /
    rows.filter(lambda x : (int(x[2])//10) == i).map(lambda x : int(x[3])).count():.1f}명
    {'*'*100}''')
    
    
```

### 선생님 코드
(간결하게 성능 고려한 코드 참고하기)

```python
lines = sc.textFile('file:///' + file_path.replace('\\','/'))
lines
```

```python
# 나이와 친구만 추출
def parseLine(line):
    fields = line.split(',')
    age = int(fields[2])
    numFriends = int(fields[3])

    return age, numFriends
```

```python
lines_rdd = lines.map(parseLine)
lines_rdd.take(3)
```

```python
age_cnt_rdd = lines_rdd.mapValues(lambda x : (x,1))
age_cnt_rdd.take(3)
```

```python
#친구 합계 구하기
age_total_cnt = age_cnt_rdd.reduceByKey(lambda x, y : (x[0]+y[0], x[1]+y[1]))
age_total_cnt.take(3)
```

```python
#평균 친구수 구하기
avge_avg_cnt = age_total_cnt.mapValues(lambda x : x[0] / x[1])
avge_avg_cnt.take(3)
```

```python
# 가장 친구가 많은 사용자 찾기
rdd2 = lines.map(lambda row : row.split(','))
rdd2.take(3)
```

```python
# 친구가 가장 많은 사용자 찾기
struct_rdd = rdd2.map(lambda x : (int(x[0]), x[1], int(x[2]), int(x[3])))
struct_rdd.take(3)
```

```python
struct_rdd.max(key = lambda x : x[3])
```

```python
# 친구 수가 100 이상인 사용자 필터링
struct_rdd.filter(lambda x : x[3]>= 100).collect()
```

```python
# 특정 나이에 속한 사람들의 친구 수 합계
age_cnt_sum = struct_rdd.map(lambda x : (x[2], x[3])).reduceByKey(lambda x, y : x+y)
age_cnt_sum.take(3)
```

```python
user_input = 30
age_cnt_sum.filter(lambda x : x[0] == user_input).collect()
```

```python
user_input = int(input('숫자 입력하세요'))
age_cnt_sum.filter(lambda x : int(x[0]) == user_input).collect()
```

### 분석 주제 추가

#### 데이터에 컬럼을 추가해보자
=> 컬럼은 기존 user_id에서 랜덤 샘플링  
=> 이러면 서로 아는 사람 관계를 지을 수 있을 거 같다

```python
import random
random_list = [random.choice(rows.map(lambda x : x[0]).collect()) for i in rows.map(lambda x : x[0]).collect()]
```

```python
df2 = []
for i,j in zip(rows.map(lambda x: x).collect(), random_list):
    tmp = list(i)
    tmp.append(j)
    df2.append(tuple(tmp))    

df2
```

```python
rdd_df = sc.parallelize(df2)
rdd_df
```

```python
rdd_df.take(10)
```

```python
rdd_df.map(lambda x : x[3]).take(10)
```

```python
.map(lambda x : x[0]).intersection(rdd_df.map(lambda x : x[3])).take(10)
```
